```python
import numpy
import matplotlib.pyplot as plt
import idpy

%matplotlib inline
```

```python
# Block Magnetization (magnetization of the first block in vertical direction)
mag = [0,1,0]

# Block Dimension
dim = [0.06,0.06,0.06]

# Position of the center of the first block
pos = [0,0,0]

# Create block
block = idpy.cassette.Block(mag, dim, pos)

# Magnetization rotation matrix 
rot = idpy.utils.rotx90p

# Number of periods of the Halbach cassette
nr_periods = 3

# Create a Halbach cassette
halbach_cassette = idpy.cassette.HalbachCassette(block, rot, nr_periods, spacing=0, N=4)
fig, ax = halbach_cassette.plot(nr_periods=1)
```

```python
def plot_fields(x, y):
    z_plot = numpy.linspace(-0.5, 1.2, 1000)
    field_x = [] 
    field_y = [] 
    field_z = []
    for z in z_plot:
        field = halbach_cassette.field([x,y,z])
        field_x.append(field[0])
        field_y.append(field[1])
        field_z.append(field[2])   
    plt.plot(z_plot, field_y, label="By", color='blue')
    plt.plot(z_plot, field_x, label="Bx", color='orange')
    plt.plot(z_plot, field_z, label="Bz", color='green')
    plt.xlabel("z [mm]")
    plt.ylabel("B [T]")
    plt.legend()
 
def show_table(x, y):
    z_table = [0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7]
    bx = []
    by = []
    bz = []
    for z in z_table:
        field = halbach_cassette.field([x,y,z])
        bx.append(field[0])
        by.append(field[1])
        bz.append(field[2])
    print("\nx: ",x)
    print("y: ", y)
    print("z: ", z_table)
    print("\nBx: ",bx)
    print("\nBy: ",by)
    print("\nBz: ",bz)
```

```python
x = 0
y = 0.08
plot_fields(x, y)
show_table(x,y)
```
